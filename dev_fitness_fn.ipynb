```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
r = 1
x, y = np.arange(-r,r,2*r/512), np.arange(-r,r,2*r/512)
xv, yv = np.meshgrid(x,y)
```

```python
# vals = 1 - ((xv/2 - 0.5)**2 + (yv)**2)
vals = np.exp(-(22*yv**2 - 1)/(-2*(xv+1)+3))
vals[xv >= 0.25] = 0

para = 8 - 2*((xv - .8)**2 +(yv-.5)**2)
cond = (xv >= 0.3) & (yv < -0.1)
vals[cond] = para[cond]
```

```python
sns.heatmap(vals)
```
